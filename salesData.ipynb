```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("Sales.csv")
total_profit=df['total_profit'].tolist()
plt.plot(total_profit,label='Month Wise Profit')
ypoints=np.array([239600,240500,204600,271100,174700,236300,
                  212000,265000,202600,237800,294200,336800])
plt.plot(ypoints,linestyle='dotted',color='b',linewidth='5',marker='o',markerfacecolor='yellow')
plt.xlabel('Month Number')
plt.ylabel('Item sold units number')
plt.legend(loc='upper right')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("Sales.csv")
total_profit=df['total_profit'].tolist()
Month=['Jan','Feb','Mar','Apr','may','June',
       'July','Aug','Sep','Oct','Nov','Dec']
total_profit=[239600,240500,204600,271100,174700,236300,
              212000,265000,202600,237800,294200,336800]
plt.axis("equal")
plt.axis()
plt.pie(total_profit,labels=Month)
plt.show()
```
